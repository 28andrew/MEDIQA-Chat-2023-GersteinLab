```python
import pandas as pd
import numpy as np
```

```python
samsum = pd.read_csv("outputs/samsum.csv")
ami = pd.read_csv("outputs/ami.csv")
```

```python
samsum.groupby(['Model Name']).mean()[["Score"]]
```

```python
error_types = ["Missing Information", "Redundant Information", "Circumstance", "Wrong Reference", "Negation", "Object", "Tense", "Modality"]
```

```python
samsum_summary = pd.DataFrame(0, index = np.arange(6), columns = ["Model Name"] + error_types)
```

```python
samsum_summary.iloc[0, samsum_summary.columns.get_loc("Model Name")] = "BART-Baseline"
samsum_summary.iloc[1, samsum_summary.columns.get_loc("Model Name")] = "BART-Cons"
samsum_summary.iloc[2, samsum_summary.columns.get_loc("Model Name")] = "Pegasus-Baseline"
samsum_summary.iloc[3, samsum_summary.columns.get_loc("Model Name")] = "Pegasus-Cons"
samsum_summary.iloc[4, samsum_summary.columns.get_loc("Model Name")] = "T5-Baseline"
samsum_summary.iloc[5, samsum_summary.columns.get_loc("Model Name")] = "T5-Cons"
```

```python
for i, row in samsum.iterrows():
    for error_type in error_types:
        if type(row[error_type]) == str:
            samsum_summary.loc[samsum_summary["Model Name"] == row["Model Name"], error_type] += 1
```
